## Az LMNA génről és annak szerepéről
Az *LMNA* génnek a *lamin* fehérjék előállításban van jelentős szerepe. A lamin fehérjéknek két nagy csoportja van, az *A* és *B*, valamint ezek alcsoportjai amik a splicing során jönnek létre, tehát az *A* alcsoportja *A* és *C* és egyéb formációi mint az *A∆10* és a *C2*. A második csoport, a *C* már a *LMNB1*, *LMNB2* gének által létrehozott fehérje verziókat tartalmazza, *B1*, *B2*, *B3*. Mivel az emberi testben *C2*, *B3* csak az ivarsejtekben fordul elő,ezért azokat most nem taglalom a továbbiakban [`[1]`](https://pubmed.ncbi.nlm.nih.gov/32456328/).

A *lamin-A/C* azomban test legtöbb sejtjében termelődik. A sorrendi különbségek miatt az *A lamin* hosszabb valamivel mint a *C lamin*. Ezek olyan fehérjék melyek struktúrális felépítésében vesznek részt a sejtnek, pontosabban, a sejtmag falának felépítésében nagyrészt [`[2]`](https://medlineplus.gov/genetics/gene/lmna/).

Megfigyelések alapján azonban a lamina hálózat módosulása a *lamin-A* típuson alapszik, sőt ez hatással van a [RARG](https://www.genecards.org/cgi-bin/carddisp.pl?gene=RARG) gének működésére is, aminek közvetlen hatása van a sejtmagokra [`[1]`](https://pubmed.ncbi.nlm.nih.gov/32456328/).

Ugyanakkor az *LMNA* által kódolt lamin a nukleáris lamina része is *(alábbi ábrán látható)*, nem csak egy fehérje, hanem a magmembrán melletti fehérjemátrix része. Mitózis során ez egy visszafordítzhatóan szétbomló folyamat. A gén mutációja több betegség kialakulásához is vezethet, mint például:  Emery-Dreifuss izomdisztrófia, familiáris részleges lipodystrophia, végtagöv izomdisztrófia, dilatációs kardiomiopátia, Charcot-Marie-Tooth-kór, Hutchinson-Gilford progéria szindróma [`[3]`](https://www.genecards.org/cgi-bin/carddisp.pl?gene=LMNA&keywords=LMNA). A továbbiakban viszont ezek közül a *Hutchinson-Gilford progéria szindrómá*t emelném ki.

## Hutchinson-Gilford progéria szindróma (HGPS)

A HGPS egy rendkívűl ritka halálos genetikai betegség, mely több klinikai betegség korai kialakulását hozza magával, melyek közül a leglátványosabb talán a gyermekek korai öregedése. Az esetek közel 80%-át a korábban bemutatott *laminA/C* okozza, pontosabban az LMNA gén 11-es exonjában létrejövő `G608G (GGC > GGT)` pontmutáció jelenik meg ami csonka *A-lamin*t hoz létre. A betegség ritka mivolta miatt gyakorlatilag szinte minden egyes beteget egyenként tanulmányoznak és elemeznek ki. Érdekességképp megemlíteném, hogy kifejezetten kevés afrikai betegről tudunk [`[1]`](https://pubmed.ncbi.nlm.nih.gov/17511383/).

A betegséget Hutchinson azonosította 1886-ban, de csak röviddel az ezredforduló után sikerült megállapítani, hogy genetikai természetű betegségről van szó. A halál beálltát többnyire érelmeszesedés szövődményei okozzák, mint a szívinfarktus, stroke, átlagosan 13 éves kor körül [`[2]`](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3230028/). Ritka esetekben hamarabb is bekövetkezhet a halál, vagy egészen 20 éves korig élehetnek. A tünetek már egy éves kor után vagy még hamarabb jelentkeznek, ezek lehetnek: aránytalan növekedés, szűkült arc, megjelennek a vénák, hajhullás, mutált hang, bőrelváltozás. Szintén megjelenik még az extrém fogyás, inzulin rezisztencia, halláskárosodás, abnormális fog növekedés, izomsorvadás [`[3]`](https://www.mayoclinic.org/diseases-conditions/progeria/symptoms-causes/syc-20356038).


> Adatokban kifejezve [`[4]`](https://www.malacards.org/card/hutchinson_gilford_progeria_syndrome):
> - várható élettartam: 10-20 év
> - általában *(median)* diagnosztizálva: 19 hónapos korban
> - 80%-ban `laminA/C`n jelenik meg, `G608G` pontmutáció okán, `C -> T` váltás


A mutáció során kialakult új láncot *Progerin*nek nevezzük, és ez egyes feltételezések szerint akár szerepet játszhat a természetes öregedésben is [`[5]`](https://www.ucsf.edu/news/2011/10/103955/aging-disease-children-sheds-light-normal-aging). De ennek kutatása a kevés esetszám (maximum 150-et feltételeznek [`[6]`](https://web.archive.org/web/20131202231251/http://www.progeria.be/?page_id=165&lang=en)) miatt ütközik nehézségekbe.

## LMNA mutációk genetikai szempontból

Az LMNA génen kialakulható lehetséges mutációk lehetnek [`[1]`](https://www.hgmd.cf.ac.uk/ac/gene.php?gene=LMNA):
- splicing
- törlés
- insertation
- point mutation
- szubsztitúció

Gyakorisági eloszlásuk az alábbi oszlopdiagrammon látható:

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

#data from https://www.hgmd.cf.ac.uk/ac/gene.php?gene=LMNA
mutations = {"nonsense": 385,"splicing": 73, "small deletions": 60, "small insertations": 26, "small indels": 7, "gross deletions": 5, "gross insertations": 5, "rearegements": 1}
plt.bar(range(len(mutations)), list(mutations.values()), align='center')
plt.xticks(range(len(mutations)), list(mutations.keys()))
plt.xticks(rotation='vertical')
plt.show()
```

Ezek számos betegség kialakulásához vezethetnek, 14 féle képpen kialakulhat HGPS és ezen kívűl még 5 féle atipikus módon is progeroid szindróma léphet fel. A legtöbb esetben egyszerű pontmutáció a HGPS okozója [`[2]`](https://pubmed.ncbi.nlm.nih.gov/12714972/). Hasonló progreoid szinrómákat mutatnak a Werner szindróma betegei is. A Werner esetén a `WRN` gén cDNS-e a 8-as kromoszómán a teleomerek, azaz a DNS végén található ismétlődő rész romlását eredményezi ami a későbbi másolatok készítésekor okoz problémát. Az ilyen szindrómában szenvedők azonban átlag 54 éves korig is elélnek, illetve érdekes módon Japánban és Sziciliában elterjedt, valamint több mint 1000 eset ismert a világon, így messze nem nevezhető olyan extrém betegségnek mint a HGPS  [`[3]`](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4140030/), ugyanakkor atipikus esetben ez is szintén az LMNA génen létrejövő mutáció [`[1]`](https://www.hgmd.cf.ac.uk/ac/gene.php?gene=LMNA).

### HGPS bioinformatikai szempontból

Mivel a HGPS nagy arányban a LMNA G608G mutációt jelenti [`[1]`](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3230028/) ezért alábbiakban csupán ennek feldolgozását szeretném megtenni. Az említett pontmutáció az `A` prelamin 50 aminosavának törlését eredményezi [`[2]`](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2092390/). Megjegyezném, hogy a HGPS ezért nem hat az agyi tevékenységre jelentősen, mivel az agyban `laminC`  van jelen dominánsan, és ritka a `laminA` [`[3]`](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3289373/), ezért feltételezhetjük, hogy nem történik demens elváltozás a betegek esetében.

Az elérhető adatbázisokban közül sajnos több esetben zárt vagy fizetős a kapcsolódó adathalmaz, pl [HGMD/LMNA](https://www.hgmd.cf.ac.uk/ac/gene.php?gene=LMNA) [`[4]`](https://ommbid.mhmedical.com/content.aspx?bookid=2709&sectionid=225070099). Ezért más adatforrásokat kellett keresnem [`[5]`](http://dx.doi.org/10.1136/jmg.2007.052639) így találtam az OMIM, GDB, Mutation Discovery forrásokat, valamint néhány másikat ami szóbajöhet, mint a [washington.edu](http://www.pathology.washington.edu/research/werner/database/), [HUGENET](http://www.hugenet.org.uk/resources/databases/table2.html), [mutdb.org](http://mutdb.org/), [UMD](http://www.umd.be/), [DMuDB](http://ngrl.man.ac.uk/dmudb/index.html) valamint a [CBiT](http://cbit.snu.ac.kr/database-frame.htm) az [amp.org-on megosztott lista](https://www.kaggle.com/datasets/sndorburian/catalogue-of-mutation-database-websites) jóvoltából [`[6]`](https://www.amp.org/AMP/assets/File/resources/mutationdatabasesummary_012809.xls). Így ezekben való keresésre koncentráltam, amelyek közül sajnos sok a halott link, vagy a valamilyen okból elérhetetlen adatbázis, még a mutdb is. 

Mivel egyedül a [sanger.ac.uk](https://cancer.sanger.ac.uk/cosmic/gene/analysis?ln=LMNA_ENST00000368301) adatbázisában találtam [eredményt](https://cancer.sanger.ac.uk/cosmic/gene/samples?all_data=&coords=AA%3AAA&dr=&end=573&gd=&id=292155&ln=LMNA_ENST00000368301&mut_pie1=CzT&seqlen=573&src=gene&start=1) de letölteni onnan sem tudtam így inkább a mutáció azonosítójával(*G608G*) kezdtem keresni a NIH adatbázisában, és itt is csak az eredeti [humán](https://www.ncbi.nlm.nih.gov/gene/4000) és [egér](https://www.ncbi.nlm.nih.gov/gene/16905) adathalmezt találtam meg a keresett pontmutáció nélkül, így inkább magam állítottam elő, tudva, hogy a 11. helyen `C>T` mutációról van szó.
