Basic Blosum calculation, with the given matrix:

> B A B A
>
> A A A C
> 
> A A C C
> 
> A A B A
> 
> A A C C 
> 
> A A B C

This is also solved as an example at [RNaseTech - Lect 4c how to create BLOSUM matrices](https://www.youtube.com/watch?v=uKWkQDB9Je4) and [BioPandit - BLOSUM Substitution Matrix](https://www.youtube.com/watch?v=ZUjAKgVrir4). The BioPython code idea is from official [biopython.org](https://biopython.org/docs/1.77/api/Bio.Align.html#Bio.Align.PairwiseAligner) tutorial.


```python
from Bio import Align
from Bio.Align import substitution_matrices
from Bio.Align import MultipleSeqAlignment
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq

aligner = Align.PairwiseAligner()
aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")
sequencies = ["BABA", "AAAC", "AACC", "AABA", "AACC", "AABC"]
for i in sequencies:
    for j in sequencies:
        alignments = aligner.align(i, j)
        alignments = list(alignments)
        print("Number of alignments: %d" % len(alignments))
        alignment = alignments[0]
        print("Score = %.1f" % alignment.score)
        print(alignment)
```
